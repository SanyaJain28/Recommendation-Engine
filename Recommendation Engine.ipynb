```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import sklearn
```

```python
movies=pd.read_csv("movies.csv")
```

```python
ratings=pd.read_csv("ratings.csv")
```

```python
movies
```

```python
ratings
```

```python
movies.drop('genres', axis=1, inplace=True)
```

```python
ratings.drop('timestamp', axis=1, inplace=True)
```

```python
merged=pd.merge(movies, ratings, on='movieId')
merged
```

```python
df=merged.dropna(axis=0)
```

```python
rating_count=df.groupby(df['movieId']).rating.count().reset_index()
rating_count.rename(columns={'rating':'total_rating'}, inplace=True)
rating_count
```

```python
new_df=pd.merge(merged, rating_count, on='movieId')
```

```python
new_df
```

```python
new_df=new_df[new_df['total_rating']>50]
new_df
```

```python
new_df.isnull().describe()
```

```python
new_df_pivot=new_df.pivot_table(index='title', columns='userId', values='rating').fillna(0)
new_df_pivot
```

```python
type(new_df)
```

```python
#convert the pivot table into matrix
from scipy.sparse import csr_matrix
rating_matrix=csr_matrix(new_df_pivot.values)
```

```python
rating_matrix
```

```python
from sklearn.neighbors import NearestNeighbors
rating_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
rating_knn.fit(rating_matrix)
```

```python
# this gives the indices of the movies that are near to target movie and its distance
#and we have selected the 90th movie randomly
distances, indices =rating_knn.kneighbors(new_df_pivot.iloc[90,:].values.reshape(1, -1), n_neighbors = 6)
#we use reshape because we do not want a series from the pivot table
```

```python
#since we selectd 6 neighbours which gives result of 5 neighbours as 1st one is the movie itself
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(new_df_pivot.index[90]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, new_df_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```
